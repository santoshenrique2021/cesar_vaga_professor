# Gráficos da aula

```python
#Bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#Eliminar os warnings
import warnings
warnings.filterwarnings("ignore")
##Ver todas as colunas do data frame
pd.set_option('display.max_columns', None)
##Ver todas as linhas do data frame
pd.set_option('display.max_rows', None)
```

# 1 - Importar os dados

```python
df = pd.read_csv('dados_vendas.csv')
#Visualização dos dados
df.head()
```

# 2 - Tipos de gráficos

## Gráfico de barras 

* Comparar os valores de uma variável categórica. Por exemplo, relaciona as **Vendas** (variável contínua) e a **Região** (variável categórica).

```python
plt.figure(figsize=(8, 4))
ax = sns.barplot(data=df, x='Regiao', y='Vendas', estimator=sum, errorbar=None)
plt.title('Vendas por Região')
ax.set_xlabel("Região")  # Renomeia o eixo x
ax.set_ylabel("Vendas Totais")  # Opcional: renomear o eixo y também
plt.tight_layout()
plt.show()
```

```python
# **Discussão**:  
# - Por que não usar um gráfico de setor (gráfico de pizza)? (Muitas regiões → barras são mais legíveis).
```

## Gráfico de dispersão 

* Comparar os valores de duas variáveis quantitativas - **Renda** e **Idade**.
* Ajuda a compreender se existe alguma relação entre as variáveis.

```python
plt.figure(figsize=(8, 4))
sns.scatterplot(data=df, x='Idade', y='Renda_Mensal')
plt.title('Renda vs Idade')
plt.show()
```

## Histograma

* Aplicado a uma variável contínua, analisando a sua distribuição.

```python
plt.figure(figsize=(8, 4))
ax = sns.histplot(data=df, x='Idade', bins=5)
ax.set_title("Distribuição de Idades")
ax.set_ylabel("Contagem")
# Garante que o eixo y mostre apenas valores inteiros
ax.set_yticks(range(int(ax.get_ylim()[1]) + 1))  # Intervalo de 0 até o máximo (arredondado)
plt.tight_layout()
plt.show()
```

## Gráfico Ruim

```python
#Pie Chart com 5 categorias (exemplo abaixo)
plt.figure(figsize=(6, 6))
plt.pie(df['Vendas'], labels=df['Categoria'], autopct='%1.1f%%')
plt.title('Proporção de Vendas (NÃO RECOMENDADO)')
plt.show()
```
